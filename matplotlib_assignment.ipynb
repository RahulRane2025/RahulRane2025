```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# --------------------------------------------
# 1. Line Plot
# --------------------------------------------
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

plt.figure(figsize=(6, 4))
plt.plot(x, y, marker='o', linestyle='-', color='blue')
plt.title("Simple Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()

# --------------------------------------------
# 2. Bar Graph
# --------------------------------------------
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

plt.figure(figsize=(6, 4))
plt.bar(students, marks, color=['red', 'green', 'blue', 'orange'])
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.ylim(0, 100)
plt.show()

# --------------------------------------------
# 3. Pie Chart
# --------------------------------------------
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

explode = [0.1 if rev == max(revenue) else 0 for rev in revenue]  # Highlight max

plt.figure(figsize=(6, 6))
plt.pie(revenue, labels=regions, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Revenue Distribution by Region")
plt.axis('equal')  # Make the pie a circle
plt.show()

# --------------------------------------------
# 4. Histogram
# --------------------------------------------
np.random.seed(0)  # For reproducibility
data = np.random.randint(1, 101, size=1000)

plt.figure(figsize=(8, 5))
plt.hist(data, bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram of Random Integers (1 to 100)")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
