```python

import numpy as np
import matplotlib.pyplot as plt

# Line Plot
# Plot the data

x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

plt.plot(x, y)
plt.show()

# plot
plt.plot(x, y, marker="o", linestyle="-", color="green")
plt.title("Line Plot of Y over X")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.show()

# Bar Graph--> Student Marks
Students = ["John", "Jane", "Alice", "Bob"]
Marks = [75, 85, 60, 90]

plt.bar(Students, Marks)
plt.show()

# Colors & Title
colors = ["Blue", "Orange", "Green", "Red"]

plt.bar(Students, Marks, color=colors)
plt.title("STUDENTS MARKS IN SUBJECT")
plt.xlabel("STUDENTS")
plt.ylabel("MARKS")
plt.ylim(0, 100)
plt.show()

# Pie Chart:Region name as labels
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
explode = [0.1 if r == max(revenue) else 0 for r in revenue]

# Pie Chart:Region with the highest revenue
plt.pie(revenue, labels=regions, autopct='%1.1f%%', startangle=140, explode=explode, shadow=True)
plt.title("COMPANY REVENUE BY REGION ")
plt.axis("EQUAL")
plt.show()

#Histogram of Random Integers

random_data = np.random.randint(1, 101, 1000)

plt.hist(random_data, bins=10, color="purple", edgecolor="black")
plt.title("Histogram of Random Integers (1–100)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
